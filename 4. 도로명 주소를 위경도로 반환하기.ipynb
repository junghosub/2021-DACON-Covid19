# 4. 도로명 주소를 위경도로 반환하기

앞서 도로명 주소를 구했고, 이를 위경도로 반환하기 위해 geocoding을 사용하였다.

```python
import pandas as pd
```

```python
# 상점 도로명 주소 및 위경도
store = pd.read_excel('data/상점 위경도.xlsx')
store
```

```python
# 목적지 도로명 주소 및 위경도
destination = pd.read_excel('data/목적지 위경도.xlsx')
destination
```

```python
# 구로구 데이터 불러오기
guro = pd.read_csv('data/구로구 - 도로명 주소.csv')
guro
```

```python
# merge를 통해  배달 상점의 위경도 Mapping
temp = pd.merge(guro, store, left_on = 'DLVR_STORE_ROAD', right_on = '상점 도로명 주소', how = 'left')

# 컬럼명 변경
temp.drop('상점 도로명 주소', axis = 1, inplace = True)
temp.rename(columns = {'Latitude' : 'DLVR_STORE_LAT'}, inplace = True)
temp.rename(columns = {'Longitude' : 'DLVR_STORE_LONG'}, inplace = True)
temp.sample(3)
```

```python
# store[store['상점 도로명 주소'] == '해당 주소']
```

```python
# merge를 통해 배달 목적지의 위경도 Mapping
temp2 = pd.merge(temp, destination, left_on = 'DLVR_DSTN_ROAD', right_on = '목적지 도로명 주소', how = 'left')

# 컬럼명 변경
temp2.drop('목적지 도로명 주소', axis = 1, inplace = True)
temp2.rename(columns = {'Latitude' : 'DLVR_DSTN_LAT'}, inplace = True)
temp2.rename(columns = {'Longitude' : 'DLVR_DSTN_LONG'}, inplace = True)
temp2.sample(3)
```

```python
# destination[destination['목적지 도로명 주소'] == '해당 주소']
```

```python
temp2.drop(['DLVR_STORE_ROAD_CODE', 'DLVR_DSTN_ROAD_CODE'], axis = 1, inplace = True)
```

```python
# csv로 저장
temp2.to_csv('data/구로구 데이터(전처리 완료).csv', index = False)
```

위에서 구한 위경도를 통해 차후 시각화 분석에서 사용하겠음.
