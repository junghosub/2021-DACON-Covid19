# 도로명 주소 쉽게 구하기

[도로명 주소 안내 시스템](https://www.juso.go.kr/openIndexPage.do)에서 제공하는 도로명 데이터를 사용할 것이다. 

```python
# 라이브러리
import pandas as pd
```

```python
# 해당 시의 도로명 정보 조회를 불러온다.
# 다만 시군구 코드는 엑셀 파일에서 추가해줘야함
guro = pd.read_excel('data/도로명정보조회(구로구).xls', header= 1)
guro.head()
```

```python
guemchoen = pd.read_excel('data/도로명정보조회(금천구).xls', header = 1)
guemchoen.head()
```

```python
yeondeung = pd.read_excel('data/도로명정보조회(영등포구).xls', header = 1)
yeondeung.head()
```

```python
gwangmyeong = pd.read_excel('data/도로명정보조회(광명시).xls', header= 1)
gwangmyeong.head()
```

```python
yangcheon = pd.read_excel('data/도로명정보조회(양천구).xls', header = 1)
yangcheon.head()
```

```python
gwanak = pd.read_excel('data/도로명정보조회(관악구).xls', header = 1)
gwanak.head()
```

```python
bucheon = pd.read_excel('data/도로명정보조회(부천시).xls', header = 1)
bucheon.head()
```

```python
dongjak = pd.read_excel('data/도로명정보조회(동작구).xls', header = 1)
dongjak.head()
```

```python
# 쉽게 하기 위해 list화 해준다.
df = [guro, guemchoen, yeondeung, yangcheon, gwanak, bucheon, dongjak, gwangmyeong]
```

```python
# for문을 통해 '도로명코드'라는 변수 생성
for row in df:
    row['시군구 코드'] = row['시군구 코드'].astype('string')
    row['도로명번호'] = row['도로명번호'].astype('string')
    row['도로명코드'] = row['시군구 코드'] + row['도로명번호']
```

```python
# for문을 통해 병합해줌.
list_of_df = [] 

for i in range(0, len(df)):
    df_temp = df[i]
    list_of_df.append(df_temp)
    
df_accum = pd.concat(list_of_df).reset_index(drop = True)
```

```python
df_accum['도로명 주소'] = df_accum['시도명'] + ' ' + df_accum['시군구명'] + ' ' + df_accum['도로명']
df_accum
```

위 데이터와 같이 본 분석에서 사용하는 도로명코드를 맞춰준 후, 이에 해당하는 도로명 주소를 구한다.

```python
df_accum.to_csv('data/도로명정보조회.csv', index = False)
```
