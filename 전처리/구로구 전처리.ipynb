# 구로구 데이터 전처리

```python
# 전처리를 위한 라이브러리
import pandas as pd
```

```python
# 데이터 불러오기
del_call = pd.read_csv('data/KT_빅데이터플랫폼_데이터셋/dlvr_call.csv')
```

```python
# 구로구만 살펴보기
guro = del_call[del_call['DLVR_DSTN_SIGNGU_NM'] == '구로구'].reset_index(drop = True)
guro = guro[guro['DLVR_REQUST_STTUS_VALUE'] == 1]
print(guro.shape)
guro.head()
```

### 배달 상품금액

```python
# 배달 상품금액이 0인 데이터
guro[guro['GOODS_AMOUNT'] == 0].reset_index(drop = True)
```

배달 상품금액이 0인 데이터가 약 3만개 가량 있다. 상식적으로 배달 상품금액이 0일 수는 없다. 기프티콘이나 쿠폰을 통해 구매한게 아닐까 싶다. 하지만 차후 분석에서 배달 금액을 이용해야하기 때문에 편의상 제외하도록 하였다.

```python
# 배달 상품금액이 0인 넘는 데이터만 보기
guro = guro[guro['GOODS_AMOUNT'] > 0]
```

```python
# 가게와 배달 목적지의 도로명 주소 코드를 str 타입으로 변경해주기
guro['DLVR_STORE_RDNMADR_CODE'] = guro['DLVR_STORE_RDNMADR_CODE'].astype('string')
guro['DLVR_DSTN_RDNMADR_CODE'] = guro['DLVR_DSTN_RDNMADR_CODE'].astype('string')
```

### 도로명 주소 변환

```python
# 배달상점도로명주소 코드
list(guro['DLVR_STORE_RDNMADR_CODE'].value_counts().index)
```

```python
# 배달상점 도로명주소 코드 변환
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.14703E+11', '114703000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.14704E+11', '114704000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15302E+11', '115302000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15303E+11', '115303000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15304E+11', '115304000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15452E+11', '115452000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15453E+11', '115453000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15454E+11', '115454000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15603E+11', '115603000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.15604E+11', '115603000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('1.16203E+11', '116203000000', inplace = True)

guro['DLVR_STORE_RDNMADR_CODE'].replace('4.11903E+11', '411903000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('4.11904E+11', '411904000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('4.11905E+11', '411905000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('4.12103E+11', '412103000000', inplace = True)
guro['DLVR_STORE_RDNMADR_CODE'].replace('4.12104E+11', '412104000000', inplace = True)
```

```python
# 배달목적지도로명주소 코드
list(guro['DLVR_DSTN_RDNMADR_CODE'].value_counts().index)
```

```python
# 배달목적지도로명주소 코드 변환
guro['DLVR_DSTN_RDNMADR_CODE'].replace('1.15302E+11', '115302000000', inplace = True)
guro['DLVR_DSTN_RDNMADR_CODE'].replace('1.15303E+11', '115303000000', inplace = True)
guro['DLVR_DSTN_RDNMADR_CODE'].replace('1.15304E+11', '115304000000', inplace = True)
guro['DLVR_DSTN_RDNMADR_CODE'].replace('1.15305E+11', '115305000000', inplace = True)
```

```python
# 이상치 
anomaly = guro[(guro['DLVR_STORE_RDNMADR_CODE'] == '\\N') | (guro['DLVR_DSTN_RDNMADR_CODE'] == '\\N')].reset_index(drop = True)
anomaly
```

```python
anomaly[['DLVR_STORE_LEGALDONG_NM', 'DLVR_STORE_RDNMADR_CODE',  'DLVR_DSTN_LEGALDONG_NM', 'DLVR_DSTN_RDNMADR_CODE']]
```

위 데이터로는 각기 다른 도로명 코드를 얻기는 힘들 것 같다. 따라서 배달상점 도로명주소, 배달목적지 도로명주소가 '\N'으로 표기된 데이터는 본 분석에서 제외하도록 하였다.

```python
# 전처리를 완료한 최종 구로구 데이터
final = guro[(guro['DLVR_STORE_RDNMADR_CODE'] != '\\N') & (guro['DLVR_DSTN_RDNMADR_CODE'] != '\\N')].reset_index(drop = True)
final
```

```python
# to_csv로 파일 저장
final.to_csv('구로구 전처리 데이터.csv', index= False)
```

### 주요 변경점
- 배달이 취소된 데이터는 제외하였음. (DLVR_REQUST_STTUS_VALUE'] == 1)
- 배달 상품금액이 0인 데이터 또한 3만개 가량 되었다. 상식적으로 배달 상품금액이 0인 것은 말이 안 된다. 기프티콘이나 쿠폰을 통해 구매한 것으로 보인다. 다만 차후 분석에서 배달금액을 통한 분석을 진행하기 때문에 제외하였다. 
- E+*로 표기된 도로명 주소를 정수로 변환하였음. 예로 1.15304E+11 -> 115304000000로 변환. 하지만 115304000000라는 도로명 번호를 가진 주소를 찾을 수 없었다.
- 배달 상점 도로명코드, 배달 목적지 도로명코드가 '\N'로 저장된 데이터가 많았다. 다른 데이터와의 연관을 통해 유추하려 했지만 불가능해보였으므로 제외하였음.
